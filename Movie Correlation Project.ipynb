### Importing all needed libraries 

```python
import pandas as pd
import numpy as np

import seaborn as sns 
import matplotlib 
import matplotlib.pyplot as plt 
plt.style.use('ggplot')

from matplotlib.pyplot import figure 
%matplotlib inline 
matplotlib.rcParams['figure.figsize']=(12,8)
```

### Reading our CSV File 

```python
df=pd.read_csv(r'C:\Users\hp\Downloads\archive\movies.csv')
df.head()
```

### Cleaning the Data

```python
# Checking if there is any missing values
for col in df.columns:
    pct_missing=np.mean(df[col].isna())   #isnull() is the same as isna()
    print('{}-{}%'.format(col,pct_missing))
```

```python
# Dropping the Null rows
df=df.dropna()
df.shape
```

```python
# Checking if there are Null Values
df.isna().any()
```

```python
# Datatypes for Columns 
df.dtypes
```

```python
# Changing the type of budget and gross

df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
df.head()
```

```python
df1=df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
df1
```

```python
pd.set_option('display.max_rows', None)
```

```python
df1.shape
```

```python
# Deleting all duplicates

df1.columns.duplicated()
```

```python
#There are no duplicates 
```

```python
# Scatter plot budget versus gross 

plt.scatter(x=df1['budget'], y=df1['gross'])
plt.title('Budget vs Gross Revenue')
plt.xlabel('Budget for the film')
plt.ylabel('Gross Revnue of the film')
plt.show()
```

```python
# Regression Plot between budget and gross
sns.regplot(x=df1['budget'], y=df1['gross'], scatter_kws={"color":"purple"},line_kws={"color":"red"}) 
```

```python
# Lets start looking at correlation 

non_numeric_columns = df1.select_dtypes(exclude=['float', 'int']).columns
numeric_df = df1.drop(columns=non_numeric_columns)
numeric_df = numeric_df.dropna()
correlation_matrix = numeric_df.corr(method='pearson')
print(correlation_matrix)


sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Metric for Numeric Features')
plt.ylabel('Movie Feature')
plt.xlabel('Budget for the Film')
plt.show()
```

```python
# Look at company -- It is not numerical but we can create numeric representations of it 


df_numerized=df1
for col_name in df_numerized.columns: 
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
df_numerized.corr()
```

```python
correlation_mat=df_numerized.corr()
corr_pairs=correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr=sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
