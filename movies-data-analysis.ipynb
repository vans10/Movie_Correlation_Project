# Importing all needed libraries 

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib 
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
from matplotlib.pyplot import figure 
%matplotlib inline 
matplotlib.rcParams['figure.figsize']=(12,8)
```

# Reading the CSV File 

```python
df=pd.read_csv('/kaggle/input/movies/movies.csv')
df.head()
```

```python
#The dimensions of our dataset
df.shape
```

# Data Cleaning

```python
#Lets see if there is any missing value 
for col in df.columns:
    pct_missing=np.mean(df[col].isna())   #isnull() is the same as isna()
    print('{}-{}%'.format(col,pct_missing))
```

```python
#Drop the Null rows
df=df.dropna()
df.shape
```

```python
# Cehcking if there are Null or NAN Values
df.isna().any()
```

```python
#Datatypes for Columns 
df.dtypes
```

```python
#Changing the type of budget and gross
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
df.head()
```

```python
df1=df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
df1
```

```python
pd.set_option('display.max_rows', None)
```

```python
df1.shape
```

```python
#Deleting all duplicates
#There are no duplicates 
df1.columns.duplicated()
```

```python
# Scatter plot budget versus gross 
plt.scatter(x=df1['budget'], y=df1['gross'])
plt.title('Budget vs Gross Earning')
plt.ylabel('Budget for film')
plt.xlabel('Gross Earning')
plt.show()
```

```python
# regplot regression plot between budget and gross
sns.regplot(x=df1['budget'], y=df1['gross'], scatter_kws={"color":"red"},line_kws={"color":"blue"}) # or x='', y='', data=df1
```

```python
#lets start looking at correlation 
df1.corr(method='pearson') #pearson , kendall, spearman
```

```python
# High correlation between budget and gross 
correlation_matrix=df1.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Metric for Numeric Features')
plt.ylabel('Movie Feature')
plt.xlabel('Budget for Film')
plt.show()
```

```python
# Look at company -- It is not numerical but we can create numeric representations of it 
df_numerized=df1
for col_name in df_numerized.columns: 
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized
        
```

```python
df_numerized.corr()
```

```python
correlation_mat=df_numerized.corr()
corr_pairs=correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr=sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
